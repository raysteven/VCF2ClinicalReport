```python
workdir = '/path/to/workdir'
raw_vcf_file = 'normal_sample.deepvariant.vcf'
intervar_file = 'normal_sample.hg38_multianno.txt.intervar'
functotator_file = 'normal_sample.deepvariant_filtered_merged_annotated.vcf'
```

```python
import sys
sys.path.insert(0, workdir)
from DataIO import loadDF, loadVCF, save_to_excel 
import os
import pandas as pd

# Adjust display options
pd.set_option('display.max_columns', None)
```

```python
import re

def extract_significance(text):
    """
    Extracts the significance description from a given text string.

    Parameters:
    text (str): The input text containing the significance description.

    Returns:
    str: The extracted significance description, or an empty string if not found.
    """
    # Regex pattern to find the desired text
    pattern = r"InterVar:\s*(.*?)\s*PVS1="
    
    # Search using the pattern
    match = re.search(pattern, text)
    
    # If a match is found, return the captured group; otherwise, return an empty string
    if match:
        return match.group(1)
    else:
        return ""
```

```python
##Read, reformatting, and add unique identifier of the InterVar Result

intervar_df = loadDF(file_path=(os.path.join(workdir, intervar_file)),file_type='txt')
intervar_df['#Chr'] = 'chr' + intervar_df['#Chr'].astype(str)
intervar_df['U_IDENTIFIER'] = intervar_df['#Chr'].astype(str) + "_" +intervar_df['Start'].astype(str)
intervar_df
```

```python
##Read, reformatting, and add unique identifier of the Funcotator Result

funcotator_df = loadVCF(file_path=(os.path.join(workdir, functotator_file)))
funcotator_df['U_IDENTIFIER'] = funcotator_df['#CHROM'].astype(str) + "_" +funcotator_df['POS'].astype(str)
funcotator_df
```

```python
## Merge the InterVar and Funcotator DF together based on common column "U_IDENTIFIER"
intervar_funcotator_df = pd.merge(funcotator_df, intervar_df, on='U_IDENTIFIER', how='inner')
intervar_funcotator_df
```

```python
## Drop All NA on the Orpha result
intervar_funcotator_df = intervar_funcotator_df.dropna(subset=['Orpha'])
intervar_funcotator_df
```

```python
## Drop SNP that has no Consequence on protein
intervar_funcotator_df = intervar_funcotator_df[intervar_funcotator_df['ExonicFunc.refGene'] != '.']
intervar_funcotator_df
```

```python
##Split Orpha columns into other dataframe to extract the information within
split_orpha_columns = intervar_funcotator_df['Orpha'].str.split('|', expand=True)

orpha_count_metadata = split_orpha_columns[1].value_counts()
orpha_count_dict = {}
for key, value in orpha_count_metadata.items():
    orpha_count_dict[key] = value

split_orpha_columns
```

```python
## Assign all the necessary information from the Orpha Result table into the main dataframe

intervar_funcotator_df['Orpha Number'] = split_orpha_columns[0]
intervar_funcotator_df['Condition Name'] = split_orpha_columns[1]
intervar_funcotator_df['Prevalence'] = split_orpha_columns[2]
intervar_funcotator_df['Inheritance Pattern'] = split_orpha_columns[3]
intervar_funcotator_df['Age of Onset'] = split_orpha_columns[4]

intervar_funcotator_df
```

```python
## Extract the ACMG Classification from the InterVar column
intervar_funcotator_df['ACMG Classification'] = intervar_funcotator_df[' InterVar: InterVar and Evidence '].apply(extract_significance)
intervar_funcotator_df
```

```python
## Mapped out the total SNP number for each "Condition Name" in this sample (Useful for filtering later on)
intervar_funcotator_df['Total Number of SNP for this condition in this sample'] = intervar_funcotator_df['Condition Name'].map(orpha_count_dict)
intervar_funcotator_df
```

```python
## Extract the ClinVar Significance from the ClinVar column and change the "UNK" to "Unknown"

intervar_funcotator_df['clinvar: Clinvar '] = intervar_funcotator_df['clinvar: Clinvar '].str.replace('clinvar:', '', regex=False)
intervar_funcotator_df['clinvar: Clinvar '] = intervar_funcotator_df['clinvar: Clinvar '].str.replace('UNK', 'Unknown', regex=False)
intervar_funcotator_df
```

```python
## Rename all the columns needed

rename_dict = {
    '#CHROM': 'Chromosome',
    'POS': 'Position',
    'REF': 'Reference Allele',
    'ALT': 'Alternate Allele',
    'QUAL': 'Variant Quality',
    'FILTER': 'Filter Status',
    'clinvar: Clinvar ':'ClinVar Significance',
    'Freq_gnomAD_genome_ALL':'gnomAD AF',
    'Freq_1000g2015aug_all' : '1000g AF',
    'ExonicFunc.refGene': 'Consequence'
}

intervar_funcotator_df = intervar_funcotator_df.rename(columns=rename_dict)
intervar_funcotator_df
```

```python
## Create report dataframe that consists of required information

to_report_column = ['Chromosome','Position','Reference Allele', 'Alternate Allele', 'Variant Quality','Filter Status','Consequence', 'ClinVar Significance','gnomAD AF','1000g AF','Condition Name','Prevalence','Inheritance Pattern','Age of Onset','ACMG Classification']
intervar_funcotator_df_to_report = intervar_funcotator_df[to_report_column]
intervar_funcotator_df_to_report
```

```python
## Save the result report into excel file

save_to_excel(file_name=(os.path.join(workdir, 'result_to_report.xlsx')), dfs={'result':intervar_funcotator_df_to_report})
```
